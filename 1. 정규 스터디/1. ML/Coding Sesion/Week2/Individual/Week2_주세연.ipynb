`````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
`````

`````python
wine = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/'
                     'wine/wine.data',header=None)
`````

`````python
X = np.array(wine.iloc[:,1:])
y = np.array(wine.iloc[:,0])
y_1 = np.where(y == 1, True, False)  
y_1
`````

`````python
# train-test set 분리 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_1, test_size=0.33, random_state=64)
`````

`````python
y_test
`````

`````python
# sklearn의 GridSearchCV 활용하기 

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

param_grid = [{'weights': ["uniform", "distance"], 'n_neighbors': [3, 4, 5]}]

knn_clf = KNeighborsClassifier()
grid_search = GridSearchCV(knn_clf, param_grid, cv=5, verbose=0, scoring='f1')
grid_search.fit(X_train,y_train)
`````

`````python
# grid search best parameter

grid_search.best_params_   
`````

`````python
# grid search best score

grid_search.best_score_
`````

`````python
# test set 의 confusion matrix

from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

pred_GS = cross_val_predict(grid_search, X_test, y_test, cv=5)
confusion_matrix(y_test,pred_GS)
`````

`````python
# test set의 f1 score

from sklearn.metrics import f1_score

f1_score(y_test,pred_GS)
`````
