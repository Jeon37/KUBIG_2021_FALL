```python
import pandas as pd
import numpy as np  
dat_wine=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/'
                     'wine/wine.data',header=None)
dat_wine.columns = ['class label', 'alchohol', 'malic acid', 'ash', 
                    'alcalinity of ash', 'magnesium', 'total phenols', 
                    'flavanoids', 'nonflavanoid phenols', 
                    'proanthocyanins', 'color intensity', 'hue', 
                    'OD280', 'proline']  # Column names
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

wine = dat_wine.drop("class label", axis=1)
wine_labels = dat_wine['class label']

X_train, X_test, y_train, y_test = train_test_split(wine,wine_labels
                                                    , test_size=0.33, random_state=64)
```

```python
y_train_1 = (y_train==1)
y_test_1 = (y_test==1)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier()
grid_params = {
    'n_neighbors' : list(range(1,20)),
    'weights' : ["uniform", "distance"],
    'metric' : ['euclidean', 'manhattan', 'minkowski']
}
grid_search = GridSearchCV(knn_clf, param_grid=grid_params, cv=5, verbose=0, scoring='f1')
```

```python
grid_search.get_params()
```

```python
grid_search.fit(X_train,y_train_1)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
final_model= grid_search.best_estimator_
```

```python
from sklearn.metrics import confusion_matrix
final_predictions = final_model.predict(X_test)
confusion_matrix(y_test_1,final_predictions )
```

```python
from sklearn.metrics import f1_score
f1_score(y_test_1,final_predictions)
```
