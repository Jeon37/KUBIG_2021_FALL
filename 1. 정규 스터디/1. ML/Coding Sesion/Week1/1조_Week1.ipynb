```python
data = pd.read_csv('../data/UNdata_Export.csv')
data.head()
```

```python
data_1=data[['Country or Area','Subgroup','Year',"Value"]] 
df=data_1.dropna()
df
```

```python
sns.barplot(data = df, x = 'Country or Area', y = 'Value', hue = 'Subgroup')
```

```python
sns.violinplot(data = df , x = 'Year', y = 'Value', hue = 'Subgroup',split = True)
plt.title('UN Life Expectancy Year & Sex',fontsize = 15)
plt.show()
```
