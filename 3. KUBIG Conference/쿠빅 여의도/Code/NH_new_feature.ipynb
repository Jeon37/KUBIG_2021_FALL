```python
# outlier 처리

def outliers_iqr(data):
  q1, q3= np.percentile(data, [25,75])
  iqr=q3-q1
  lower_bound=q1-(iqr*5)
  upper_bound=q3+(iqr*5)

  return np.where((data>upper_bound)|(data<lower_bound))
```

```python
과거보유일_outlier=outliers_iqr(train_selected_1['과거 보유일'])[0]
잔고금액_outlier=outliers_iqr(train_selected_1['잔고금액'])[0]
수익률_outlier=outliers_iqr(train_selected_1['수익률'])[0]

outlier_index = np.concatenate((과거보유일_outlier, 잔고금액_outlier,수익률_outlier), axis=None)

print(len(outlier_index))
```

```python
not_oulier_index=[]

for i in train_selected_1.index:
  if i not in outlier_index:
    not_oulier_index.append(i)

train_clean_1=train_selected_1.loc[not_oulier_index]
train_clean_1=train_clean_1.reset_index(drop=True)
```
