## KNN Collaborative filtering

```python
hold_matrix=inv_data_test.pivot(index='계좌ID', columns='종목코드', values='보유기간(일)')
```

```python
hold_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
matrix_dummy=hold_matrix.copy().fillna(0)
user_similarity=cosine_similarity(matrix_dummy, matrix_dummy)
user_similarity=pd.DataFrame(user_similarity, index=hold_matrix.index, columns=hold_matrix.index)
```

```python
def cf_knn(user_id, inv_code, neighbor_size=0):
    if inv_code in hold_matrix:
        sim_scores = user_similarity[user_id].copy()
        hold = hold_matrix[inv_code].copy()
        none_rating_idx = hold[hold.isnull()].index
        hold = hold.drop(none_rating_idx)
        sim_scores = sim_scores.drop(none_rating_idx)
   
        if neighbor_size == 0:          
            mean_hold = np.dot(sim_scores, hold) / sim_scores.sum()

        else:                       
            if len(sim_scores) > 1: 
                neighbor_size = min(neighbor_size, len(sim_scores))
                sim_scores = np.array(sim_scores)
                hold = np.array(hold)
                user_idx = np.argsort(sim_scores)
                sim_scores = sim_scores[user_idx][-neighbor_size:]
                hold = hold[user_idx][-neighbor_size:]
                mean_hold = np.dot(sim_scores, hold) / sim_scores.sum()
            else:
                mean_hold = 10
    else:
        mean_hold = 10
    return mean_hold
```
